## Introduction
---

A major characteristic that distinguishes Eukaryotes from Prokaryotes is the presence of introns in the Eukaryotic genome. According to the National Human Genome Research Institute (USA), an <b>intron</b> "is a region that resides within a gene but does not remain in the final mature mRNA molecule following transcription of that gene and does not code for amino acids that make up the protein encoded by that gene"[1]. Introns are considered a burden on the eukaryotic genomes and are considered more deleterious than adaptive [2]. It takes a lot of energy to read, copy and excise introns at the correct positions [3]. In fact, many genetic disorders result from splicing mistakes like ALS, Myotonic dystrophy, Spinal muscular atrophy etc [2]. With this information in mind, it seems reasonable to think that introns should not have propagated along lineages. 

Yet its interesting to see that some lineages like <i>Saccharomyces cerevisiae, Drosophila melanogaster and Caenorhabditis elegans</i> have experienced an extensive loss of introns whereas some lineages like primates have continued to propagate introns [3]. It cannot be ignored that there would be more genetic drift acting on smaller, simpler, unicellular eukaryotes than larger, complex, multicellular exukaryotes due to their population sizes [2]. Despite this is it possible that introns may just serve a beneficial. Is an intron always a burden? Maybe not?

While introns increase mutability and affect fitness, it may also act as a buffer against mutations. For Group II and Group I intron the spliceosomes identify the 5′ splice site (CAG/GUAAGU sequence) and 3′ splice site (NYAG/G sequence) to splice the intron [4]. Thus, any mutation that occurs in the intron is not likely to result in the frameshift of the coding sequence (CDS). So, it should be that the more introns there are in a gene the less likely it is for a frameshift mutation to progagate to the whole CDS. Therefore, acting as a potential buffer.



Many studies, have looked into and found direct and indirect functionalimportance of introns like [3]: 
    <ol>
        <li>Introns enhancing gene expression </li>
        <li>Intron sequences control mRNA transport or chromatin assembly</li>
        <li>Introns in 5' and 3' UTR affect non-sense mediated decay  </li>
        <li>Different ordinal positions of introns may have different functional roles</li>
        <li>Intron length matters in the efficiency of natural selection</li>
    </ol> 

In this study, we want to examine two major relationships: 
    <ol>
        <li>The relationship between Number of Introns in a transcript and the Conservation Scores of the Coding Sequence and the Untranslated Regions (UTRs)</li>
        <li>The relationship between Length of the Upstream Intron and the Conservation Score of the Coding Sequence of the Downstream Exon</li>
    </ol>

Since, some eukaryotic lineages have excessively loss introns like <i>Saccharomyces cerevisiae, Drosophila melanogaster and Caenorhabditis elegans</i>. Whereas some have passed them on like <i>Homo sapiens, Mus musculus</i>. Thus, it would be interesting to see how the above mentioned relationships differ between these species.

Let's flatten out the 9th column of the GTF to make it easier to sort. So after processing, the transcripts GTF file will have 12 columns and the exons GTF file will have 13 columns. All separated by tabs -'\t'

```python
def unique_fields(old_gtf:str):
    old_file_ptr = open(old_gtf, "r")
    unique_fields = {}
    counter = 0
    
    lines = old_file_ptr.readlines()
    for line in lines:
        columns = line.split("\t")
        fields = columns[8].strip("\n").split(";") # Column 9
        for field in fields:
            if field != "":
                parts = field.strip(" ").split(" ")
                if parts[0] not in unique_fields:
                    unique_fields[parts[0]] = counter
                    counter += 1
    old_file_ptr.close()
    return unique_fields
    
def flatten_gtf(old_gtf: str, new_gtf: str): 
    found_fields = unique_fields(old_gtf)
    extras = len(found_fields)
    old_file_ptr = open(old_gtf, "r")
    new_file_ptr = open(new_gtf, "w")
    
    lines = old_file_ptr.readlines()
    for line in lines:
        columns = line.split("\t")
        fields = columns[8].strip("\n").split(";") # Column 9
        values = ['<NA>'] * extras
        for field in fields:
            if field != "":
                parts = field.strip(" ").split(" ")
                index = found_fields[parts[0]]
                values[index] = parts[1].strip('\"')

        # Write the existing        
        new_file_ptr.write("\t".join(columns[:8]) + "\t")
        # Write the flattened values to new GTF
        new_file_ptr.write("\t".join(values) + "\n") # "\n" ends the current line
        
    new_file_ptr.close()
    old_file_ptr.close()

flatten_gtf("exons/sacCer3.all.exons.gtf", "exons/sacCer3.flattened.exons.gtf")
flatten_gtf("transcripts/sacCer3.all.transcripts.gtf", "transcripts/sacCer3.flattened.transcripts.gtf")
flatten_gtf("exons/ce11.all.exons.gtf", "exons/ce11.flattened.exons.gtf")
flatten_gtf("transcripts/ce11.all.transcripts.gtf", "transcripts/ce11.flattened.transcripts.gtf")
flatten_gtf("exons/dm6.all.exons.gtf", "exons/dm6.flattened.exons.gtf")
flatten_gtf("transcripts/dm6.all.transcripts.gtf", "transcripts/dm6.flattened.transcripts.gtf")
flatten_gtf("exons/mm10.all.exons.gtf", "exons/mm10.flattened.exons.gtf")
flatten_gtf("transcripts/mm10.all.transcripts.gtf", "transcripts/mm10.flattened.transcripts.gtf")
flatten_gtf("exons/hg38.all.exons.gtf", "exons/hg38.flattened.exons.gtf")
flatten_gtf("transcripts/hg38.all.transcripts.gtf", "transcripts/hg38.flattened.transcripts.gtf")
```

```python
cons_ind_map = {0: '0 < CS <= 0.4', 1: '0.4 < CS <= 0.6', 2: '0.8 < CS <= 1'}
species_cons = []
cons_lst = []
# Collecting all the data for Lost Group
for parser in parsers:
    total = 0
    dist = [0,0,0]
    sp = []
    for gene in parser:
        for transcript in gene:
            if transcript.avg_cds_cons > 0: # -1 not allowed
                total += 1
                if transcript.avg_cds_cons <= 0.4:
                    dist[0] += 1 
                elif transcript.avg_cds_cons <= 0.6:
                    dist[1] += 1
                else: 
                    dist[2] += 1 
                sp.append(transcript.avg_cds_cons)                
    species_cons.append(sp)
    if total > 0:
        for i in range(len(dist)):
            dist[i] /= total   
            cons_lst.append({'Species': parser.species, 'Percentage': dist[i], 'Conservation Level': cons_ind_map[i]})

cons_df = pd.DataFrame(cons_lst)
```

Let's see what proportion of the coding sequences (CDS) in the genes have a Conservation Score (CS) between (0,0.4], (0.4,0.6] and (0.6,1]. It will be displayed as a ratio CS (0,0.4] : CS (0.4,0.6] : CS (0.6,1] .

```python
species = ['sacCer3', 'ce11', 'dm6', 'mm10', 'hg38']
totals = [0,0,0]
for sp in species:
    subset_df = cons_df[cons_df['Species'] == sp]
    ratios = subset_df['Percentage'].values
    print("Ratios in {} - {:.2f} : {:.2f} : {:.2f}".format(sp, ratios[0], ratios[1], ratios[2]))
    totals[0] += ratios[0]
    totals[1] += ratios[1]
    totals[2] += ratios[2]
print("Average Ratio {:.2f} : {:.2f} : {:.2f}".format(totals[0]/5,totals[1]/5,totals[2]/5))
```

For all the species (with exception of ce11), regardless of the Group, there is a hike in the CDS Conservation Score when number of introns increase from 0 to 2, after that the conservation seems pretty consistent. Let's take a look if there is a significant difference in conservation levels between 0 to 2 introns. Let's do an ANOVA to find out.

<b>Null hypothesis</b> - there is no significant difference in conservation levels between having 0 - 2 introns.<br>
<b>Alternate hypothesis</b> - there is a significant difference in conservation levels between 0 - 2 introns.<br>

We will reject the null hypothesis if p-value < 0.05

```python
sacCer3 = pd.DataFrame(in_cds['sacCer3'])
sacCer3_2 = sacCer3[sacCer3['Number']==2]['Conservation']
sacCer3_1 = sacCer3[sacCer3['Number']==1]['Conservation']
sacCer3_0 = sacCer3[sacCer3['Number']==0]['Conservation']
stat, p_value = f_oneway(sacCer3_2,sacCer3_1, sacCer3_0)
print("sacCer3 statistic: {:.4f} p-value: {:.4f}".format(stat, p_value))

ce11 = pd.DataFrame(in_cds['ce11'])
ce11_2 = ce11[ce11['Number'] == 2]['Conservation']
ce11_1 = ce11[ce11['Number'] == 1]['Conservation']
ce11_0 = ce11[ce11['Number'] == 0]['Conservation']
stat, p_value = f_oneway(ce11_2, ce11_1, ce11_0)
print("ce11 statistic: {:.4f} p-value: {:.4f}".format(stat, p_value))

dm6 = pd.DataFrame(in_cds['dm6'])
dm6_2 = dm6[dm6['Number'] == 2]['Conservation']
dm6_1 = dm6[dm6['Number'] == 1]['Conservation']
dm6_0 = dm6[dm6['Number'] == 0]['Conservation']
stat, p_value = f_oneway(dm6_2,dm6_1, dm6_0)
print("dm6 statistic: {:.4f} p-value: {:.4f}".format(stat, p_value))

mm10 = pd.DataFrame(in_cds['mm10'])
mm10_2 = mm10[mm10['Number'] == 2]['Conservation']
mm10_1 = mm10[mm10['Number'] == 1]['Conservation']
mm10_0 = mm10[mm10['Number'] == 0]['Conservation']
stat, p_value = f_oneway(mm10_2, mm10_1, mm10_0)
print("mm10 statistic: {:.4f} p-value: {:.4f}".format(stat, p_value))

hg38 = pd.DataFrame(in_cds['hg38'])
hg38_2 = hg38[hg38['Number'] == 2]['Conservation']
hg38_1 = hg38[hg38['Number'] == 1]['Conservation']
hg38_0 = hg38[hg38['Number'] == 0]['Conservation']
stat, p_value = f_oneway(hg38_2, hg38_1, hg38_0)
print("hg38 statistic: {:.4f} p-value: {:.4f}".format(stat, p_value))
```

Based on the values of the ANOVA test, it seems that for the Gained Group the difference in Conservation Scores from 0 introns to 2 introns is statistically significant indicating that indeed the conservation increases as the number of introns increase from 0 to 2. The ANOVA test values for the Lost Group are not as high as those from the Gained Group except dm6. Indicating that the difference may not be as significant as it was in Gained Group. But in both cases, an increase in number of introns from 0 to 2 results in an increase in conservation score (exception ce11).

```python
dist = { 'sacCer3': {'VS': 1, 'S': 62, 'M': 252, 'L': 435, 'VL': 505},
         'ce11': {'VS': 45, 'S': 50, 'M': 301, 'L': 895, 'VL': 100912},
         'dm6' : {'VS':58, 'S': 65, 'M': 498, 'L': 3518, 'VL': 268107},
        'mm10' : {'VS':162, 'S': 659, 'M': 3016, 'L': 10644, 'VL': 1067346},
        'hg38' : {'VS':195, 'S': 804, 'M': 3955, 'L': 14545, 'VL': 1160411}
       }
ind_map = {'VS': 0, 'S': 1, 'M': 2, 'L': 3, 'VL': 4}
il_cds = []
for parser in parsers:
    for gene in parser:     
        for transcript in gene:
            if not transcript.intronless:
                intron_lens = transcript.get_intron_lengths()               
                for c in range(0,len(transcript.cds_cons)-1): # First exon does not have 
                    if intron_lens[c] <= dist[parser.species]['VS']:
                        il_cds.append({'Type': 'VS', 'Value': ind_map['VS'],'Conservation': transcript.cds_cons[c+1]})
                    elif intron_lens[c] <= dist[parser.species]['S']:
                        il_cds.append({'Type': 'S', 'Value': ind_map['S'], 'Conservation': transcript.cds_cons[c+1]})
                    elif intron_lens[c] <= dist[parser.species]['M']:
                        il_cds.append({'Type': 'M','Value': ind_map['M'], 'Conservation': transcript.cds_cons[c+1]})
                    elif intron_lens[c] <= dist[parser.species]['L']:
                        il_cds.append({'Type': 'L', 'Value': ind_map['L'],'Conservation': transcript.cds_cons[c+1]})
                    else:
                        il_cds.append({'Type': 'VL', 'Value': ind_map['VL'],'Conservation': transcript.cds_cons[c+1]})

cds = pd.DataFrame(il_cds)
cds_cut = cds[cds['Conservation'] > 0] 

#Correlation Processing
df_pearson_corr, df_p_value = pearsonr(cds_cut['Value'], cds_cut['Conservation'])
print("\nPearson Correlation co-efficient between Intron Length and Conservation Score is {:.4f} with p-value of {:.4f}".format(df_pearson_corr,df_p_value))

# Finding the proportions of the conservations at each location
props = []
groups = [0,0.4,0.6,1.0]
for i in dist['dm6'].keys():
    y_values = []
    for j in range(len(groups)-1):
        total = cds_cut[cds_cut['Type'] == i].shape[0]
        found = cds_cut[(cds_cut['Type'] == i) & (cds_cut['Conservation'] > groups[j]) & (cds_cut['Conservation'] <= groups[j+1])].shape[0]
        y_values.append(found/total * 100)
    props.append(y_values)
stacked = pd.DataFrame(props, columns=['0 < CS <= 0.4', '0.4 < CS <= 0.6', '0.6 < CS <= 1'], index=dist['dm6'].keys())
print("\nPercentage of CDS with Different Levels of Conservation for Intron Lengths")
print(stacked)
```

## Discussion
---
From the introduction, we known that the species in the Lost Group have excessively lost introns and the species from the Gained Group has gained more and more introns and passed them along. If introns acted as a buffer towards mutations, then then species that have more introns would have more coding sequences with higher conservation scores. The relationship between the number of introns and the conservation score (CS) of the coding sequence (CDS) was examined for all 5 species (sacCer3, ce11, dm6, mm10 and hg38) (Figure 2). For all the species (exception of ce11), a statistically significant hike in the CDS conservation was seen when the number of introns increased from 0 to 2. It must be noted that the hike was more statistically significant for the species in the Gained Group compared to thosed in the Lost Group (with the exception of dm6). This sudden increase in conservation scores indicates that introns may indeed be involved in protecting the downstream exons against potential frameshift mutations. There is yet to understand why the increase in conservation score stabilizes after the second intron. Why does adding more introns not contribute to the conservation score? It must be clarified that the data only suggests a correlational relationship not a causation relationship. We also conducted an comparative analysis on the distribution of conservation levels of CDS in the genomes across all the species (Figure 1). It was found that regardless of the number of introns a species had, the ration of the conservation levels remained consistent.

Secondly, we analyzed the relationship between the number of introns and the conservation of the 3' UTR and the 5' UTR (Figure 3). Only 3 species had  UTRs (hg38, mm10 and dm6). For the three, only dm6 showed a significant weak positive correlation. For this species the 3' UTR was more conserved than the 5' UTR for the same number of introns. This makes sense because 3' UTR are downstream of introns whereas not all 5' UTRs have introns upstream of them. Apart from this observation, the relationship between intron number and conservation of UTRs does not seem to be correlational. It could be because there are other factors TIS (Transcription Initiation Site), PCS (Premature Start Codons) etc that influence UTR conservation. Lastly, we analyzed the effect of length of the upstream intron on the conservation of the CDS of the downstream exon. Again, a very weak pearson correlation was found amongst the two, but it must be noted that the proportion of CDS's that had a conservation score between 0.6 - 1 increased as the relative length of the intron increased.

To conclude, it seems that it is 1st and 2nd intron that are important to CDS conservation and that introns may play a role in buffering against mutations. A previous study has found a link between the expression levels of a gene and the conservation of the first intron [5]. Thus, the first and second introns seems to play an important role in gene regulation. For future studies, I would recommend analyzing the coding sequence (amino acids) of the 1st and 2nd introns to see if they share something in common that can help us understand why they are so important.
